# Last but not least, lets get inside clusters!

In this tutorial, we will demonstrate the difference in smooth and cliff clusters. Smooth clusters are devoid of activity cliffs, but cliff clusters most likely do contain some amount of ACs. For this demonstration, we consider the first cluster of CHEMBL2034_Ki dataset. You can change the SMILES file accordinggly but ensure that you change the data and property prefix while loading.

```python
# Load the input data 

smiles_file='data/CHEMBL2034_Ki.csv'  # Ensure this is consistent with the npy files

from scripts.smooth_ac import FingerprintClusterAnalyzer

#Intialize the analyzer
analyzer = FingerprintClusterAnalyzer(
        fingerprint_types=['ECFP', 'MACCS', 'RDKIT'],
        thresholds=np.linspace(0.3,0.9,7), # Adjust thresholds as needed
        top=10
    )

# Load data and perform clustering
print("Loading fingerprint data...")
analyzer.load_fingerprint_data(data_prefix='files/fps_CHEMBL2034_Ki_fp_', prop_prefix='files/props_CHEMBL2034_Ki_fp_') # Change file paths accordingly.

print("\nPerforming clustering...")
analyzer.perform_clustering()

# Save clustering results
analyzer.save_results_to_csv(filename='results/clustering_results_CHEMBL2034_Ki.csv')

# Compare fingerprint types at a specific threshold
print("\n" + "="*80)
print("EXAMPLE 2: Fingerprint Type Comparison")
print("="*80)
analyzer.compare_fingerprint_types(0.9, 0, smiles_file, max_molecules=20) # Visualizes and saves the first (or most populated cluster) across all fingerprint types at 0.9 threshold.
```
